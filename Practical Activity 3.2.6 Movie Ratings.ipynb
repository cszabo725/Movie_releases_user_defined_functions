```python
# Import the required packages
import pandas as pd
import numpy as np

# Create the DataFrames
movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')

# Merge the DataFrames with a left join 
mov_ott = movies.merge(ott, how='left', on='ID') 

# View the mov_ott DataFrame
mov_ott
```

```python
# Create new DataFrame 
mov_ott_ratings = mov_ott[['Rotten Tomatoes', 'ID', 'IMDb']]

# View the DataFrame
print(mov_ott_ratings)
```

```python
# Replace NaN values with 0 in IMDb and Rotten Tomatoes
mov_ott_ratings_final = mov_ott_ratings.fillna(0)

mov_ott_ratings_final
```

```python
# Add a new column to the DataFrame indicating average rating
# Average rating is ((IMDb/10)+ Rotten Tomatoes)/2
# Write a user defined function
def av_col2(df1, df2):
    df = ((df1/10)+df2)/2
    return(df)

mov_ott_ratings_final['average_ratings'] = av_col2(mov_ott_ratings_final['IMDb'],
                                             mov_ott_ratings_final['Rotten Tomatoes'])

mov_ott_ratings_final
```

```python
# How many movies were released per content rating (age)
def cat_count(df1):
    print(df1.value_counts())

# Number of movies released per 'Age'
df = mov_ott['Age'].astype('category')

# View the output
cat_count(df)
```

```python
# How many movies were released per year?
def year_count(df1):
    print(df1.value_counts())
    
# Number of movies released per year
df = mov_ott['Year'].astype('category')

# View the output
year_count(df)
```
